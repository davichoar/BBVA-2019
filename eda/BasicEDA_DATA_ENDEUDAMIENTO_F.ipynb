### Rangos, distribuci√≥n

Rangos en los tipos de deuda. Tener en cuenta que algunos pueden ser covariables. Evaluaremos. Al parecer tienen una distribuci√≥n uniforme. No poseen valores nulos.

```python
range_cols = list(set(df_endeudamiento.columns) - set(['MES', 'ID_CLIENTE', 'CD_BANCO', 'ST_CREDITO']))
df_endeudamiento_ranges = df_endeudamiento[range_cols]
df_endeudamiento_ranges.head()
```

```python
# Heatmap
plt.figure(figsize=(18,10))
sns.heatmap(df_endeudamiento_ranges.corr(),annot=True)
```

### Observaciones importantes:
- RGO_SDAVALES y RGO_SDLEA_PEM poseen una correlaci√≥n de 1 (Rango de importe de deuda en Avales y Rango de importe de deuda total sin considerar leasing ni Pestamo comercial).
- Fuerte correlaci√≥n entre RGO_SD_SINLEASING (Rango de importe de deuda total sin considerar leasing) y 2 variables: RGO_SDDIRECTO (Rango de importe de deuda en productos como prestamos personal, tarjeta de credito y credito vehicular) y RGO_SDTOTAL (Rango de importe de deuda total en el sistema financiero)
