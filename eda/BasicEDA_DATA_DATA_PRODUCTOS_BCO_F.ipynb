No hay nulos. Los features son de numero de cuentas respecto a un tipo de cuentas, y el rango de importe.

```python
#Columnas para el heatmap
heat_cols = list(set(df_productos.columns) - set(['MES', 'ID_CLIENTE']))
```

```python
# Heatmap
plt.figure(figsize=(42,24))
sns.heatmap(df_productos[heat_cols].corr(),annot=True, linewidth=0.5)
```

```python
corr_vars = df_productos[heat_cols].corr()

for col in heat_cols:
    #Correlation with output variable
    cor_target = abs(corr_vars[col])
    #Selecting highly correlated features
    relevant_features = cor_target[cor_target>0.5]
    print('For column',col)
    print
    print(relevant_features)
```
